```python
class Solution:
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1

            elif arr[mid] == 1:
                mid += 1

            else:  # arr[mid] == 2
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr


# Example test run
if __name__ == "__main__":
    s = Solution()
    print(s.sort012([0, 1, 2, 0, 1, 2]))
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1

        while low <= high:
            mid = (low + high) // 2

            if arr[mid] == k:
                result = mid       # possible smallest index
                high = mid - 1     # search left side

            elif arr[mid] < k:
                low = mid + 1

            else:  # arr[mid] > k
                high = mid - 1

        return result


# Example test run
if __name__ == "__main__":
    s = Solution()
    print(s.binarysearch([1, 2, 3, 4, 5], 4))
```
